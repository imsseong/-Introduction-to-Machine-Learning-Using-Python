### 2. 선(line)으로 시각화하기

#### 1) 1차원 데이터 생성 및 시각화
- python 패키지 가져오기, matplotlib 출력 옵션 설정

```python
import matplotlib as mpl
import matplotlib.pyplot as plt # 서브패키지인 pyplot, 플롯 그려주는 패키지
import numpy as np              # 행렬,벡터 등 수학계산 위한 자료구조와 계산함수 제공


%matplotlib inline
#ipytion의 magic function 중 하나로, 
# matplot의 시각화 결과를 ipytion notebook안에서 출력
```

- 임의의(랜덤) 값 40개로 이루어진 배열 변수 생성  
numpy.random : 랜던값들을 생성하는 여러 함수를 모아놓은 패키지  
numpy.random.standard.normal() : 표준 정규 분포를 따르는 랜덤 값을 생성하는 함수

```python
value = np.random.standard_normal(40)
value # array 형태로 40개의 난수값
```

- 선 plot 그리기 및 x축,y축 범위 설정

```python
index = range(len(value)) # range형 변수는 연속적인 값을 갖는 자료형
plt.plot(index, value)
plt.xlim(0, 20)
plt.ylim(np.min(value) - 1, np.max(value) + 1)
```

- plot 옵션 설정 및 꾸미기

```python
plt.figure(figsize=(7,4)) # 7과 4라는 값을 튜플 값으로 설정, 플롯의 새로운 모양 생성
plt.plot(value.cumsum(), 'b', lw=1.5) # 표시할 색깔 'b' 블루
plt.plot(value.cumsum(), 'ro') # 표시할 스타일 'ro' 빨간원
plt.xlabel('index') # x축 이름
plt.ylabel('value') # y축 이름
plt.title("Line Plot 1") # 플롯 제목
```

#### 2) 2차원 데이터 생성 및 시각화

- 랜던 값으로 이루어진 2차원 배열 변수 생성하기

```python
value = np.random.standard_normal((30,2)) # 30행 2열의 배열
value
```

- 한 그래프에 그리기

```python
plt.figure(figsize=(10, 5)) # 그림의 크기 설정
plt.plot(value[:, 0], lw = 1.5, label = '1st')
plt.plot(value[:, 1], lw = 1.5, label = "2nd")
plt.plot(value, 'ro')
plt.grid(True)
plt.legend(loc = 0)
plt.xlabel('index')
plt.ylabel('value')
plt.title('Line Plot 2')
```

하나의 플롯 그래프 안에 두 개의 데이터 셋이 독립적으로 그려짐  
각각의 데이터 셋에 라벨을 붙여 구분

- 데이터 별로 각각 그리기 : subplot() 함수 이용

```python
plt.figure(figsize=(10,5))
plt.subplot(211) # 2행 1열 서브플롯중 하나
plt.plot(value[:, 0], lw = 1.5, label = '1st')
plt.plot(value[:, 0], 'ro')
plt.grid(True)
plt.legend(loc = 0)
plt.ylabel('value')
plt.title('Line Plot 3')

# 2번째
plt.subplot(212)
plt.plot(value[:,1], lw = 1.5, label = '2nd')
plt.plot(value[:,1], '*')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.ylabel('value')
```

### 3. 산점도(scatter plot)으로 시각화하기

#### 1) pyplot.plot() 함수를 이용한 산점도 그리기
- 랜덤 값들로 이루어진 2차원 배열 변수 생성 후 plot 그리기

```python
value = np.random.standard_normal((500, 2)) # 표준정규분포 형태의 난수 500개씩  2줄
plt.plot(value[:,0], value[:,1], 'ro')
plt.grid(False)
plt.xlabel('value1')
plt.ylabel('value2')
plt.title('Scatter Plot1')
```

#### 2) pyplot.scatter() 함수를 이용한 산점도 그리기
- plot 그리기

```python
plt.figure(figsize=(7, 5))
plt.scatter(value[:,0], value[:,1], marker='o')
plt.grid(True)
plt.xlabel('value1')
plt.ylabel('value2')
plt.title('Scatter Plot2')
```

- 산점도에 색깔 입히기(c 옵션 사용)

```python
color = np.random.randint(0, 10, len(value))
plt.figure(figsize=(10, 5))
plt.scatter(value[:,0], value[:,1], c=color, marker='o')
#plt.colorbar() # 플롯에 색깔막대기 추가하는 함수
plt.xlabel('value1')
plt.ylabel('value2')
plt.title('Scatter Plot 3')
```

### 4. 히스토그램(histogram)으로 시각화하기
히스토그램 : 도수분포를 그래프로 나타낸 것, 데이터 형상과 산포정도 평가하는데 많이 사용  
히스토그램 bin : 연속적인 변수를 히스토그램으로 나타내기 위해 데이터를 나누는 개수

#### 1) pyplot.hist() 함수를 이용한 히스토그램 그리기

- 히스토그램 그리기

```python
plt.figure(figsize=(12, 7))
plt.hist(value, label=('1st', '2nd'), bins = 10)
plt.grid(True)
#오른쪽 상단의 위치로 상대적 위치값 x축 1.2 y축 1
plt.legend(bbox_to_anchor=(1.2, 1), loc='upper right') 
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram1')
```

- 히스토그램 색 입히기 - 데이터 셋 별로 다른 색 입히기

```python
plt.figure(figsize=(12, 7))
plt.hist(value, label=('1st', '2nd'), bins = 5, color = ('green', 'yellow'))
plt.grid(True)
#오른쪽 상단의 위치로 상대적 위치값 x축 1.2 y축 1
plt.legend(bbox_to_anchor=(1.2, 1), loc='upper right') 
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram1')
```

- 히스토그램 색 입히기 - 각 patch에 다른 색 입히기  
for문과 setP() 함수를 이용하자

```python
value = np.random.standard_normal(500)
# colormap 인스턴스를 가져오는 함수, Spectral이름의 컬러맵을 cm 변수로 할당
cm = plt.cm.get_cmap('Spectral')
plt.figure(figsize=(12, 6))
n, bins, patches = plt.hist(value, bins=20, color='orange')
print("1. bins : " + str(bins))
print("2. the length of bins : " + str(len(bins)))
bin_centers = 0.5 * (bins[:-1] + bins[1:])
# 파이썬의 인덱싱 함수 사용
# 인덱싱 함수 : 변수명 오른쪽에 대괄호, 그 안에 콜론과 양 옆에 인덱스 번호로
# 접근하려는 요소들의 시작점과 끝점 지정
# 시작점 공백 : 첫 번째 인덱스 부터 시작
# 끝점 공백 : 마지막까지 검색
# 인덱스 번호(-) : 배열의 뒤에서부터 접근
# 인덱스 번호(-1) : 마지막 요소 전까지
print("3. bin_centers : " + str(bin_centers))

# scale values to interval [0,1]
col = bin_centers - min(bin_centers)
col = col / max(col)
print("4. col : " + str(col))

for c, p in zip(col, patches):
    plt.setp(p, 'facecolor', cm(c))

plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram 3')
plt.show()
```

### 5. Seaborn으로 시각화하기
seaborn : matplotlib 기반으로 하는 python 시각화 패키지  
통계적 그래픽을 제공하는 고차원의 인터페이스 제공


```python
import seaborn as sns
sns.set(style='whitegrid', palette='pastel')

data = sns.load_dataset("tips")
data
```

```python
sns.boxplot(x='day', y='tip', hue='sex', data=data, palette='muted')
```

#### 1) flights 데이터로 heatmap 그리기
heatmap : 색상으로 표현할 수 있는 정보를 열 분포 형태로 나타낸 그래프
- Seaborn 내장 데이터셋 fights 로드

```python
data2 = sns.load_dataset("flights")
data2
```

- 데이터 pivot 시키기

```python
pivoted_data = data2.pivot('year', 'month', 'passengers') # index, column, value
pivoted_data
```

- 스타일 옵션 변경하여 heatmap으로 나타내기

```python
sns.set(context="poster", font="monospace")
# annot : 히트맵의 각 셀에 데이터 값 표시할 것인지, fmd d는 10진수 형태
sns.heatmap(pivoted_data, annot=True, fmt= "d", linewidth=3)
```

- 옵션 초기화시키기

```python
sns.set()
sns.reset_orig() # reset params to defaults
sns.heatmap(pivoted_data, annot=True, fmt= "d", linewidth=3)
```

### 6. Pandas로 시각화하기
#### 1) 산점도 매트릭스 그리기

```python
import pandas as pd
from pandas import scatter_matrix
value = np.random.rand(500, 4)
df = pd.DataFrame(value, columns = ['value', 'value2', 'value3', 'value4'])
df
```

```python
# df : 시각화할 데이터
# alpha : 투명도 설정
# figsize : 그림 사이즈 설정
# diagonal : 시각화 종류 설정 -> value 1,2,3,4 를 대각선에다 그려라
scatter_matrix(df, alpha=0.2, figsize=(6,6), diagonal='hist')
```
